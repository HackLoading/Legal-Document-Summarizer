# Legal Summarization Webapp - Local Run

This notebook sets up and runs the legal text summarization webapp on your local system using Streamlit. It assumes you have the webapp directory with `app.py`, `requirements.txt`, and a `.venv` virtual environment.

## Prerequisites
- Python 3.8+ installed.
- The webapp folder contains the necessary files.
- If `.venv` is not set up, run `python -m venv .venv` first.

```python
# 1) Check Python and Install Dependencies
import sys
import subprocess

print('Python version:', sys.version)

# Install requirements if not already installed
try:
    import streamlit
    import transformers
    print('Dependencies already installed.')
except ImportError:
    print('Installing dependencies...')
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-r', 'requirements.txt'])
    print('Dependencies installed.')
```

```python
# 2) Activate Virtual Environment (if using .venv)
import os

# Assuming .venv is in the webapp directory
venv_path = os.path.join('.', '.venv', 'Scripts', 'activate') if os.name == 'nt' else os.path.join('.', '.venv', 'bin', 'activate')

if os.path.exists(venv_path):
    print('Activating virtual environment...')
    !source {venv_path}  # For Unix-like systems; adjust for Windows if needed
    print('Virtual environment activated.')
else:
    print('Virtual environment not found. Ensure .venv exists or install dependencies globally.')
```
